### Maximum Subarray
Given an array of integers nums, find the subarray with the largest sum and return the sum.

A subarray is a contiguous non-empty sequence of elements within an array.



Example 1:

Input: nums = [2,-3,4,-2,2,1,-1,4]

Output: 8

Explanation: The subarray [4,-2,2,1,-1,4] has the largest sum 8.

2. Write a program using a brute-force approach to get the maximum-subarray of your student number (`40 points`)

```python
def max_subarray_brute_force(numbers):

  max_overallsum = 0

  for first in range(len(numbers)):

    max_currentsum = 0
    for last in range(first, len(numbers)):

      max_currentsum = max_currentsum + numbers[last]
    if max_currentsum > max_overallsum:

      max_overallsum = max_currentsum

  return max_overallsum

student_number = [2, 0, 2, 2, 1, 0, 1, 5, 0, 6]
subarray_sum = max_subarray_brute_force(student_number)

print("The maximum subarray sum of my student number is: ", subarray_sum)
```

3. Write a program using a dynamic programming to get the maximum-subarray of your student number (`50 points`)

```python
def max_subarray_student_num(numbers):

  max_overallsum = numbers[0]
  max_currentsum = numbers[0]

  for num in numbers[1:]:

    max_currentsum = max(num, max_currentsum + num)
    if max_currentsum > max_overallsum:

      max_overallsum = max_currentsum

  return max_overallsum

student_number = [2, 0, 2, 2, 1, 0, 1, 5, 0, 6]
subarray_sum = max_subarray_student_num(student_number)

print("The maximum subarray sum of my student number is: ", subarray_sum)
```
